```python
a=data1['Apparent Temperature_C']
b=data1['Humidity']
test=stats.ttest_rel(a,b)
pvalue = test[1]
print(pvalue)
```

```python
if pvalue>0.05:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```
